```python
gender_counts = df['Athlete gender'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Cinsiyete Göre Katılım Dağılımı')
plt.show()
```

```python
#yaş kategorilerine ggöre sporcu sayısı

plt.figure(figsize=(10, 6))
df['Athlete age category'].value_counts().plot(kind='bar', color='teal')
plt.title('Yaş Kategorilerine Göre Sporcu Sayısı')
plt.xlabel('Yaş Kategorisi')
plt.ylabel('Sporcu Sayısı')
plt.xticks(rotation=45)
plt.show()
```

```python
#Ülkelere Göre Katılımcı Sayısı (Horizontal Bar Chart)

top_countries = df['Athlete country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='barh', color='coral')
plt.title('En Fazla Katılımcı Gönderen Ülkeler')
plt.xlabel('Katılımcı Sayısı')
plt.ylabel('Ülke')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Event number of finishers'], bins=15, color='red', alpha=0.6)
plt.title('Etkinliği Bitiren Katılımcı Sayısı Dağılımı')
plt.xlabel('Bitiren Katılımcı Sayısı')
plt.ylabel('Frekans')
plt.show()

```

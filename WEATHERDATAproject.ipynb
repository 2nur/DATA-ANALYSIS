```python
# Tarih/saat sütununun ismini belirt
date_column = "Date/Time"
# Tarih/saat sütununu pandas to_datetime fonksiyonu ile dönüştür, saati kaldır
df[date_column] = pd.to_datetime(df[date_column], format='%m/%d/%Y %H:%M').dt.date
```

```python
# Nem Oranı Dağılımı (Histogram)

plt.figure(figsize=(8,6))
plt.hist(df['Rel Hum_%'], bins=20, color='purple', edgecolor='black')
plt.title('Relative Humidity Distribution')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Rüzgar Hızı ve Görüş Mesafesi İlişkisi (Scatter Plot)
plt.figure(figsize=(8,6))
plt.scatter(df['Wind Speed_km/h'], df['Visibility_km'], color='green')
plt.title('Wind Speed vs. Visibility')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Visibility (km)')
plt.show()
```

```python
# Basınç Dağılımı (KDE Plot - Yoğunluk Grafiği)

plt.figure(figsize=(8,6))
sns.kdeplot(df['Press_kPa'], shade=True, color='cyan')
plt.title('Pressure Distribution')
plt.xlabel('Pressure (kPa)')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grafiğin boyutunu ayarla
plt.figure(figsize=(12,6))

# Weather Condition sütununa göre countplot oluştur (dikey)
sns.countplot(x='Weather Condition', data=df, order=df['Weather Condition'].value_counts().index)

# Başlık ve etiketleri ayarla
plt.title('Frequency of Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Count')

# X eksenindeki etiketlerin görünürlüğünü artırmak için döndürme
plt.xticks(rotation=90)

# Grafiği göster
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grafiğin boyutunu ayarla
plt.figure(figsize=(12,6))

# Histogram grafiği oluştur
sns.histplot(df['Temp_C'], bins=30, kde=True, color='skyblue')

# Başlık ve etiketleri ayarla
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')

# Grafiği göster
plt.show()
```

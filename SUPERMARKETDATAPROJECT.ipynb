```python
# Sayısal değişkenler için fonksiyon tanımı
def sayısal_grafik(değerler):
    # Sayısal değişkenin dağılımını çizdir
    plt.figure(figsize=(9, 3))
    plt.hist(df[değerler], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel(değerler)
    plt.ylabel("Frekans")
    plt.title(f"{değerler} Değişkeninin Dağılımı")
    plt.show()

# Sayısal değişkenler listesi
sayısal = ["Sales", "Postal Code"]

# Her bir sayısal değişkenin grafiğini oluştur
for sutun in sayısal:
    sayısal_grafik(sutun)
```

```python
# Fonksiyon tanımı
def satış_grafiği(değerler):
    # Her kategori için satış toplamlarını hesapla
    cat_sales = df.groupby(değerler)["Sales"].sum()
    
    # Grafiği oluştur
    plt.figure(figsize=(9, 3))
    plt.bar(cat_sales.index, cat_sales)
    plt.xticks(cat_sales.index, cat_sales.index.values, rotation=90)
    plt.ylabel("Toplam Satış Miktarı")
    plt.title(f"{değerler} Bazında Satış Miktarları")
    plt.show()

# Kategoriler listesi
kategori = ["Region", "Category", "Segment", "Country", "State"]

# Kategori listesindeki her sütunu Sales değişkenine göre görselleştir
for sutun in kategori:
    satış_grafiği(sutun)
```

```python
#barplot kullanımı

# İlk 20 eyaletin satışlarını hesaplama
total_stage_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=True).head(20)

# Görselleştirme
plt.figure(figsize=(12, 8))
sns.barplot(data=total_stage_sales.reset_index(), x='Sales', y='State', palette='Blues')
plt.title('Top 20 States vs Sales', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.xticks(rotation=45)  # X eksenindeki etiketleri döndür
plt.tight_layout()  # Grafik alanını optimize et
plt.show()
```

```python
#yatay şekilde
#barh kullanımı

# İlk 20 eyaletin satışlarını hesaplama
total_stage_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(20)

# Görselleştirme
plt.figure(figsize=(12, 8))
plt.barh(total_stage_sales.index, total_stage_sales.values, color='skyblue')  # Yatay çubuk grafiği
plt.title('Top 20 States vs Sales', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.xticks(rotation=45)  # X eksenindeki etiketleri döndür
plt.tight_layout()  # Grafik alanını optimize et
plt.show()
```
